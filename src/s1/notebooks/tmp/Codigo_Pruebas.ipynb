```python
import re
# Asumimos que si el mensaje es saliente, el nombre del cliente está en 'A Contacto', si es entrante, está en 'Desde Contacto'
detalle_mensajes_df['Nombre_Completo'] = detalle_mensajes_df.apply(
    lambda row: row['A Contacto '] if row['Tipo Mensaje'].strip().lower() == 'saliente' else row['Desde Contacto '],
    axis=1
)

def despersonalizar_mensaje_full(contenido, nombre_completo):
    # Asegurar que ambos sean cadenas de texto y verificar que el nombre esté en el contenido
    if pd.notnull(nombre_completo) and isinstance(contenido, str):
        # Comprobar si el nombre completo está en el contenido
        if nombre_completo.lower() in contenido.lower():
            # Reemplazar el nombre completo en el contenido con un marcador genérico
            contenido = re.sub(re.escape(str(nombre_completo)), "[NOMBRE_COMPLETO]", contenido, flags=re.IGNORECASE)
    return contenido

# Aplicar la función de despersonalización a la columna de contenido
detalle_mensajes_df['Contenido_Despersonalizado'] = detalle_mensajes_df.apply(
    lambda row: despersonalizar_mensaje_full(row['Contenido '], row['Nombre_Completo']),
    axis=1
)
# Supongamos que 'df' es tu DataFrame y 'columna' es la columna que quieres analizar
def verificar_y_reemplazar(valor):
    # Definir el texto que queremos verificar
    texto_objetivo = "¡En Tigo Business impulsamos tu negocio!"
    
    # Verificar si el valor empieza con el texto objetivo
    if isinstance(valor, str) and valor.startswith(texto_objetivo):
        # Si empieza con el texto objetivo, devolver solo ese texto
        return texto_objetivo
    else:
        # Si no, devolver el valor original
        return valor

# Aplicar la función a la columna específica del DataFrame
detalle_mensajes_df['Contenido_Despersonalizado'] = detalle_mensajes_df['Contenido_Despersonalizado'].apply(verificar_y_reemplazar)
detalle_mensajes_df=detalle_mensajes_df.sort_values(['Caso ID '])

detalle_mensajes_df
```

```python
import pandas as pd
import re

# Suponiendo que el dataframe 'data' ya tiene las columnas limpias y el contenido mapeado
# Asegurarse de que la columna Fecha sea de tipo datetime
data['Fecha'] = pd.to_datetime(data['Fecha'])

# Ordenar el dataframe por 'Caso ID' y 'Fecha'
data = data.sort_values(by=['Caso ID', 'Fecha'])

# Función para obtener el último nodo numérico de los salientes
def ultimo_nodo_numerico_salida(pasos_salientes):
    # Filtrar los nodos que son numéricos
    nodos_numericos = [nodo for nodo in pasos_salientes if nodo.isdigit()]
    return nodos_numericos[-1] if len(nodos_numericos) > 0 else None

def ultimo_nodo_total (pasos_totales):
    #retornar valores ultimos del nodo total 
    return pasos_totales[-1] if len(pasos_totales) > 0 else None
# Agrupar por 'Caso ID' y crear las nuevas columnas
agrupado = data.groupby('Caso ID').agg(
    Nombre_Completo=('Nombre_Completo', 'first'),
    Pasos_Totales=('Contenido Mapeado', lambda x: list(x)),
    Pasos_Salientes=('Contenido Mapeado', lambda x: list(x[data['Tipo Mensaje'] == 'Saliente'])),
    Pasos_Entrantes=('Contenido Mapeado', lambda x: list(x[data['Tipo Mensaje'] == 'Entrante']))
)

# Agregar la columna con el último nodo numérico alcanzado en los mensajes salientes
agrupado['Ultimo_Nodo_Saliente'] = agrupado['Pasos_Salientes'].apply(ultimo_nodo_numerico_salida)
agrupado['tamaño_lista_total']=agrupado['Pasos_Totales'].apply(len)
agrupado['Ultimo_nodo']=agrupado['Pasos_Totales'].apply(ultimo_nodo_total)
# Exportar a CSV
agrupado.to_csv('listado.csv')

agrupado
```

```python
import pandas as pd

# Inicializamos una lista para almacenar los IDs de cada caso
caso_ids = []

# Definimos un diccionario para almacenar los intentos entre pares de nodos
intentos_entre_pares = {f'N{i}-N{i+1}': [] for i in range(1, 6)}  # 'N1-N2', 'N2-N3', ..., 'N5-N6'

# Iteramos sobre cada vector de nodos en el DataFrame
for index, nodos in enumerate(agrupado['Pasos_Salientes']):
    caso_id = agrupado.index[index]  # Asumimos que 'agrupado' tiene 'Caso ID' como índice
    caso_ids.append(caso_id)
    
    nodos = [str(nodo) for nodo in nodos]  # Aseguramos que todos los nodos sean str

    # Inicializamos un diccionario para almacenar los intentos entre nodos para cada caso individualmente
    intentos_por_caso = {f'N{i}-N{i+1}': None for i in range(1, 6)}  # Inicializamos con None

    # Iteramos para buscar transiciones entre nodos
    for i in range(1, 6):  # De 1 a 5
        nodo_actual = str(i)
        nodo_siguiente = str(i + 1)

        intentos = 0
        transition_found = False

        # Iteramos sobre el vector de nodos para buscar transiciones
        for j in range(len(nodos) - 1):
            if nodos[j] == nodo_actual:
                for k in range(j + 1, len(nodos)):
                    intentos += 1
                    if nodos[k] == nodo_siguiente:
                        intentos_por_caso[f'N{nodo_actual}-N{nodo_siguiente}'] = intentos 
                        transition_found = True
                        break
                    else:
                        intentos_por_caso[f'N{nodo_actual}-N{nodo_siguiente}'] = intentos
                if transition_found:
                    break
        
        # Verificar si el intento fue registrado correctamente
        print(f"Caso: {nodos}, Nodo: N{nodo_actual}-N{nodo_siguiente}, Intentos: {intentos_por_caso[f'N{nodo_actual}-N{nodo_siguiente}']}")

    # Añadimos los intentos de este caso a nuestro diccionario general
    for key, value in intentos_por_caso.items():
        intentos_entre_pares[key].append(value)

# Convertimos el diccionario a un DataFrame para mostrar los resultados
df_intentos = pd.DataFrame(intentos_entre_pares)

# Función para calcular el promedio ignorando los valores None
def mean_ignore_none(series):
    valid_values = series.dropna()  # Eliminar valores None
    if len(valid_values) > 0:
        return valid_values.mean()
    else:
        return 0  # Puedes devolver 0 o None según prefieras

# Aplicar la función a cada columna del DataFrame
promedios_por_nodo = df_intentos.apply(mean_ignore_none)

# Convertir los promedios a un DataFrame y asignar un índice
df_promedios = pd.DataFrame(promedios_por_nodo, columns=['Promedio de Intentos'])
df_promedios.to_csv('promedios_entre_nodos.csv')

# Mostrar los resultados
df_promedios
```
