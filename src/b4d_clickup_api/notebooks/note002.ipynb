```python
list_id = 901108276268
headers = {'Authorization': 'pk_81504116_WSDP2TYT7JOIM90HAUDLY4HTDTWHLUQ0'}

url = f'https://api.clickup.com/api/v2/list/{list_id}/task'
params = {'archived': False, 'page': 0}


def find_value(custom_fields, field_name: str):
    try:
        for field in custom_fields:
            if field.get('name') == field_name:
                type_config_list = field.get('type_config')
                selected_option = field.get('value')
                options = type_config_list.get('options')
                if options is not None: 
                    for option in options:
                        if option.get('orderindex') == selected_option:
                            return option.get('name')
        return None
    except Exception as e:
        print(f"Error from find_value: {e}")
        return None
    
def get_value(custom_fields, field_name: str):
    try:
        for field in custom_fields:
            if (field.get('name') == field_name) & (field.get('value') is not None):
                return field.get('value') 
        return None
    except Exception as e:
        print(f"Error from find_value: {e}")
        return None

def make_request(url, params=None):
    """
    Make a request to ClickUp API.
    """
    try:
        response = requests.get(url, headers=headers, params=params)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        #logger.error(f"Error en la solicitud a {url}: {e}")
        return None
        
def main():
    while True:
        data = make_request(url, params=params)
        task = []
        if not data:
            break

        tasks_data = data.get('tasks', [])
        for task_data in tasks_data:
            # Convert assignees json to list
            data_assignees = task_data.get('assignees')
            if isinstance(data_assignees, list):
                assignees_val = [assignee.get('username') for assignee in data_assignees if assignee.get('username')]
            elif isinstance(data_assignees, dict):
                assignees_val = [data_assignees.get('username')] if data_assignees.get('username') else []
            # Get estimated mrc value and stage
            custom_fields = task_data.get('custom_fields')
            estimated_mrc_value = get_value(custom_fields, "Estimated MRC Value")
            estimated_otc_value = get_value(custom_fields, "Estimated OTC Value")
            stage = find_value(custom_fields, "Stage")
            country = find_value(custom_fields, "Pais")
            tipo = find_value(custom_fields, "Tipo")

            # get space id
            space_id = task_data['space']['id']

            task.append({
                    'task_id': task_data['id'],
                    'name': task_data['name'],
                    'assignees': assignees_val,
                    'stage': stage,
                    'estimated_mrc_value': estimated_mrc_value,
                    'estimated_otc_value': estimated_otc_value,
                    'country': country,
                    'tipo': tipo,
                    'space_id': space_id,
            })

        if data.get('next_page'):
            params['page'] += 1
        else:
            break
    print(task)

main()
```
